# Determinants

## 1.  Basic Techniques

Let $A$ be an $n\times n$ matrix, a square matrix. The determinant of $A$, denoted by $\mathrm{det}(A)$ is a very important number which we will explore as follows. 

#### Outcomes
- Evaluate the determinant of a square matrix using either Laplace Expansion or row operations.
- Demonstrate the effects that row operations have on determinants.
- Verify the following:
    - The determinant of a product of matrices is the product of the determinants.
    - The determinant of a matrix is equal to the determinant of its transpose.

**Definition 1.** (Determinant of a Two By Two Matrix)

Let $A=\left[\begin{array}{ll}a & b \\ c & d\end{array}\right]$. Then

$$\Large 
\operatorname{det}(A)=a d-c b
$$




**Definition  2.**  (ម៉ាទ្រីសម៉ៃនរ - the $ij$th minor of a matrix)

Let $A$ be a $3 \times 3$ matrix. The **$i j^{\text{th}}$ minor of $A$**, denoted as $\operatorname{minor}(A)_{i j}$ or $M_{ij}$, is the determinant of the $2 \times 2$ matrix which results from **deleting the $i^{\text{th}}$ row and the $j^{\text {th }}$ column of $A$**.

In general, if $A$ is an $n \times n$ matrix, then the $i j^{t h}$ minor of $A$ is the determinant of the $n-1 \times n-1$ matrix which results from deleting the $i^{t h}$ row and the $j^{t h}$ column of $A$.

***Example.***  Finding Minors of a Matrix

Let

$$
A=\left[\begin{array}{lll}
1 & 2 & 3 \\
4 & 3 & 2 \\
3 & 2 & 1
\end{array}\right]
$$


Find $\operatorname{minor}(A)_{12}$ and $\operatorname{minor}(A)_{23}$.


**Definition 3.** The $i j^{\text {th }}$ Cofactor of a Matrix

Suppose $A$ is an $n \times n$ matrix. The $i j^{t h}$ cofactor, denoted by $\operatorname{cof}(A)_{i j}$ is defined to be

$$\Large 
\operatorname{cof}(A)_{i j}=(-1)^{i+j} \operatorname{minor}(A)_{i j}
$$

#### Proposition 2.1

- The **number of $r$-permutations of $n$ objects** is given by

$$
P(n, r)=n(n-1) \cdots(n-r+1)
$$

- The **number of $r$-combinations of $n$ objects** is given by

$$
\binom{n}{r}=\frac{n!}{r!(n-r)!}
$$

- The **number of permutations of an $n$-multiset of type $\left(r_1, \ldots, r_k\right)$** is the same as the *number of ways to partition an $n$-set into $k$ subsets of cardinalities $r_1, \ldots, r_k$*, and is given by

$$
\binom{n}{i_1, \ldots, i_k}=\frac{n!}{r_{1}!\cdots r_{k}!}
$$

- The **number of $n$-combinations of $r$ objects with repetition allowed** equals the number of non-negative integer solutions of $x_1+\cdots+x_r=n$, and is given by

$$
\left\langle\begin{array}{l}
r \\
n
\end{array}\right\rangle=\binom{n+r-1}{n}
$$


#### Proposition 2.2.

(1) **The Pascal identity**: $$\binom{n}{r}=\binom{n-1}{r}+\binom{n-1}{r-1}$$

(2) **An recurrence relation**: $$\binom{n+1}{r+1}=\sum_{k=r}^n\binom{k}{r}$$

(3) **The Vandermonde convolution**: $$\binom{m+n}{r}=\sum_{i=0}^r\binom{m}{i}\binom{n}{r-i}$$

(4) **The binomial expansion**: $$(x+y)^n=\sum_{i=0}^n\binom{n}{i} x^i y^{n-i}$$

(5) **The multinomial expansion**: $$\left(x_1+\cdots+x_k\right)^n = \sum_{\stackrel{i_1+\cdots+i_k=n,}{i_1,\ldots, i_k \geq 0}}\binom{n}{i_1, \ldots, i_k} x_1^{i_1} \cdots x_k^{i_k}$$

# 2. Counting of Functions

Given sets $A$ and $B$, we have the following classes of functions from $A$ to $B$: 

- $\operatorname{Map}(A, B) =\Bigl\{f: A \rightarrow B\Bigr\}$
  
- $\operatorname{Inj}(A, B) =\Bigl\{f: A \rightarrow B \mid f \text { is injective }\Bigr\}$

  
- $\operatorname{Sur}(A, B) =\Bigl\{f: A \rightarrow B \mid f \text { is surjective }\Bigr\}$

  
- $\operatorname{Bij}(A, B) =\Bigl\{f: A \rightarrow B \mid f \text { is bijective }\Bigr\}$

**Note.** If $A, B$ are linearly ordered sets, we say that a function $f: A \rightarrow B$ is monotonic provided that $x \leq y$ in $A$ implies $f(x) \leq f(y)$ in $B$. We have the class of functions

$$
\operatorname{Mon}(A, B)=\Bigl\{f: A \rightarrow B \mid f \text { is monotonic }\Bigr\}.
$$

#### Proposition 2.4. 

យក $A$ និង $B$ ជាសំណុំរាប់អស់ (finite sets) មានកាឌីណាល់ $|A|=m$ និង $|B|=n$ ។ គេបាន៖ 

<ol type="i">
<li> ចំនួនអនុគមន៍ យកពី $A\longrightarrow B$ កំណត់ដោយ៖ 
    $$\Bigl|\operatorname{Map}(A, B)\Bigr|=n^m$$
</li>

<li> ចំនួនអនុគមន៍ប្រកាន់ យកពី $A\longrightarrow B$ កំណត់ដោយ៖ 
    $$\Bigl|\operatorname{Inj}(A, B)\Bigr|=n(n-1) \cdots(n-m+1)$$
</li>

<li> ចំនួនអនុគមន៍ពេញ យកពី $A\longrightarrow B$ កំណត់ដោយ៖ 
    $$\Bigl|\operatorname{Sur}(A, B)\Bigr|=\sum_{k=0}^n(-1)^{n-k}\binom{n}{k} k^m$$
</li>

<li> ចំនួនអនុគមន៍មួយទល់មួយ យកពី $A\longrightarrow B$ កំណត់ដោយ៖ 
    $$\Bigl|\operatorname{Bij}(A, B)\Bigr|=\left\{\begin{array}{lll}n! & \text { if } \quad m=n, \\ 0 & \text { if } \quad m \neq n .\end{array}\right.$$
</li>
</ol>

#### Definition 4. 

- The falling factorial of length $n$ is

$$
[x]_{(n)}=x(x-1) \cdots(x-n+1), \quad n \geq 1
$$

and $[x]_{(0)}=1$. 

- The rising factorial of length $n$ is expression

$$
[x]^{(n)}=x(x+1) \cdots(x+n-1), \quad n \geq 1
$$

and $[x]^{(0)}=1$.


#### Proposition 2.5. (Reciprocity Law) 

For integers $n \geq 1$,

$$
\begin{aligned}
& {[-x]_{(n)}=(-1)^n[x]^{(n)},} \\
& {[-x]^{(n)}=(-1)^n[x]_{(n)} .}
\end{aligned}
$$


#### Proposition 2.6. 
Let $A$ and $B$ be linearly ordered finite sets with cardinalities $|A|=n$ and $|B|=m$. Then

$$
\Bigl|\operatorname{Mon}(A, B)\Bigr|=\dfrac{[m]^{(n)}}{n!}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 400)
y = np.where(x != 1, (x**2 - 1) / (x - 1), np.nan)  # Avoid division by zero

plt.figure(figsize=(10, 6))
plt.plot(x, y, label="f(x)=(x² - 1)/(x - 1)", color='blue')
plt.scatter([1], [2], color='red', label='Hole at (1, 2)', zorder=5)
plt.axvline(x=1, color='gray', linestyle='--', label='x=1')
plt.title("Removable Discontinuity")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.ylim(-10, 10)
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt  
import numpy as np  

# Define the function  
def f(x):  
    return np.exp(-x**2) + 2*x - 1 
  
# Create data  
x = np.linspace(0, 10, 100)  
y1 = np.sin(x)  
y2 = f(x) 

# Plot the function  
plt.figure()  
plt.plot(x, y1)  
plt.title("Sine Wave")  
plt.xlabel("X")  
plt.ylabel("Y")  

# Render the plot in the browser  
plt.show()  

plt.figure()  
plt.plot(x, y2)  
plt.title("Parabola")  
plt.xlabel("X")  
plt.ylabel("Y") 

# Render the plot in the browser  
plt.show() 
```
